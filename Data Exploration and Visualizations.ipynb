#### View Manhattan Distributions

```python
#These Distribution plots were inspired by the Stackover Question: 
#http://stackoverflow.com/questions/6620471/fitting-empirical-distribution-to-theoretical-ones-with-scipy-python

size = len(manhattan['Log_Sale_Price'].values)
x = scipy.arange(size)
y = manhattan['Log_Sale_Price'].values
h = plt.hist(y, color='w')

dist_names = ['gamma', 'beta', 'rayleigh', 'norm', 'pareto']

for dist_name in dist_names:
    dist = getattr(scipy.stats, dist_name)
    param = dist.fit(y)
    print dist_name, "parameters: ", param
    pdf_fitted = dist.pdf(x, *param[:-2], loc=param[-2], scale=param[-1]) * size
    plt.plot(pdf_fitted, label=dist_name)
    upperb = max(manhattan['Log_Sale_Price'].values)
    lowerb = min(manhattan['Log_Sale_Price'].values)
    plt.xlim(lowerb,upperb)
plt.legend(loc='upper right')

plt.xlabel('Log of Sale Price ($)')
plt.ylabel('Frequency')
plt.title('Which distribution most accurately represents the data?')
sns.despine()
plt.show()
```

#### Check to see if any zipcode has 0% of a certain race.

## Visualizations

### Look at sales by month

```python
Months = cleandf.groupby('MONTH').count()['LOT']
```

```python
Months.plot(kind='bar')
labels = ['Jan','Feb','Mar','Apr','May','June','July','Aug','Sept','Oct','Nov','Dec']
plt.title('Sales by Month')
plt.xticks([i+.05 for i in range(13)],labels)
plt.ylabel('Frequency')
plt.xlim(-.5,12.5)
sns.despine()
plt.show()
```

### Look at sales by year.

```python
Years = cleandf.groupby('YEAR').count()['LOT']
Years.plot(kind='bar')
plt.title('Sales by Year')
plt.ylabel('Frequency')
sns.despine()
plt.show()
```

### Look at buildings by year built.

```python
plt.hist((cleandf['YEAR BUILT']),bins=40)
plt.title('When were the buildings built?')
plt.xlabel('Year')
plt.ylabel('Frequency')
plt.xlim(1890,2015)
sns.despine()
plt.show()
```
